```python
import pandas as pd
import numpy as np
from sklearn.grid_search import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
```

```python
#Lectura de archivo y dummies
data = pd.read_csv('titanic_train.csv')
data = data.dropna()
data
```

```python
X_age = data['Age'].values
y_age = data['Survived'].values
X_age = X_age.reshape(-1,1)
```

```python
#RandomForest
rf = RandomForestClassifier()
#Busqueda de los parametros
n_estimators = list(range(1,31))
max_depth = list(range(1,31))
criterion = ['gini','entropy']
param_dist = dict(criterion = criterion, max_depth=max_depth,
                 n_estimators=n_estimators)
rand = RandomizedSearchCV(rf,param_dist,cv=10,scoring='accuracy',
                         n_iter=10,random_state=5)
rand.fit(X_age,y_age)
print(rand.grid_scores_)
print(rand.best_score_)
print(rand.best_params_)
```

```python
data_dummies = pd.get_dummies(data)
data_dummies.dropna
features = data_dummies.loc[:,['Age','Sex_female','Sex_male','Pclass']]
X = features.values
y = data_dummies['Survived'].values
print(list(data_dummies.columns))
```

```python
#RandomForest
rf = RandomForestClassifier()
#Busqueda de los parametros
n_estimators = list(range(1,31))
max_depth = list(range(1,31))
criterion = ['gini','entropy']
param_dist = dict(criterion = criterion, max_depth=max_depth,
                 n_estimators=n_estimators)
rand = RandomizedSearchCV(rf,param_dist,cv=10,scoring='accuracy',
                         n_iter=10,random_state=5)
rand.fit(X,y)
print(rand.grid_scores_)
print(rand.best_score_)
print(rand.best_params_)
```

```python
#Feature nueva Family Size
data_dummies["Family_Size"] = (data_dummies['SibSp'] + data_dummies['Parch'] + 1)
print(list(data_dummies.columns))
features = data_dummies.loc[:,['Age','Sex_female','Sex_male','Pclass','Family_Size']]
X = features.values
y = data_dummies['Survived'].values
#RandomForest
rf = RandomForestClassifier()
#Busqueda de los parametros
n_estimators = list(range(1,31))
max_depth = list(range(1,31))
criterion = ['gini','entropy']
param_dist = dict(criterion = criterion, max_depth=max_depth,
                 n_estimators=n_estimators)
rand = RandomizedSearchCV(rf,param_dist,cv=10,scoring='accuracy',
                         n_iter=10,random_state=5)
rand.fit(X,y)
print(rand.grid_scores_)
print(rand.best_score_)
print(rand.best_params_)
```
